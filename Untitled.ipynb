```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader
from sklearn.preprocessing import MinMaxScaler
from pathlib import Path


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

# ====================================================
# 1. Load your engineered feature dataframe
#    Assume df already contains columns listed above
# ====================================================

# ==== 路径 ====
datadir = Path("../data")
data_file = datadir / "dataset_taskA.csv"

# ==== 读取数据 ====
df = pd.read_csv(data_file, parse_dates=["timestamp"], index_col="timestamp")

FEATURES = [
    c for c in df.columns 
    if c not in ["CO2", "target_next1h"]
]

TARGET = "target_next1h"

print("Using features:", FEATURES)

# ====================================================
# 2. Scaling
#    Scale all input features and target separately
# ====================================================
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

X_all = scaler_X.fit_transform(df[FEATURES].values)
y_all = scaler_y.fit_transform(df[[TARGET]]).reshape(-1)

# ====================================================
# 3. Train / Val / Test split by time
# ====================================================
n = len(df)
i_train_end = int(n * 0.70)
i_val_end   = int(n * 0.85)

X_train_raw = X_all[:i_train_end]
y_train_raw = y_all[:i_train_end]

X_val_raw   = X_all[i_train_end:i_val_end]
y_val_raw   = y_all[i_train_end:i_val_end]

X_test_raw  = X_all[i_val_end:]
y_test_raw  = y_all[i_val_end:]

print("Train:", X_train_raw.shape)
print("Val:",   X_val_raw.shape)
print("Test:",  X_test_raw.shape)

# ====================================================
# 4. Convert to sequences
# ====================================================
SEQ_LEN = 72   # 72 hours history

def create_sequences(X, y, seq_len=72):
    xs, ys = [], []
    for i in range(len(X) - seq_len):
        xs.append(X[i:i+seq_len])        # (72, D)
        ys.append(y[i+seq_len])          # scalar
    return np.array(xs), np.array(ys)

X_train_seq, y_train_seq = create_sequences(X_train_raw, y_train_raw, SEQ_LEN)
X_val_seq,   y_val_seq   = create_sequences(X_val_raw,   y_val_raw,   SEQ_LEN)
X_test_seq,  y_test_seq  = create_sequences(X_test_raw,  y_test_raw,  SEQ_LEN)

print("X_train_seq:", X_train_seq.shape)  # expect: (N, 72, D)
print("Num features (D):", X_train_seq.shape[2])

# ====================================================
# 5. Convert to tensors and DataLoaders
# ====================================================
batch_size = 256

train_ds = TensorDataset(
    torch.tensor(X_train_seq, dtype=torch.float32),
    torch.tensor(y_train_seq, dtype=torch.float32)
)
val_ds = TensorDataset(
    torch.tensor(X_val_seq, dtype=torch.float32),
    torch.tensor(y_val_seq, dtype=torch.float32)
)
test_ds = TensorDataset(
    torch.tensor(X_test_seq, dtype=torch.float32),
    torch.tensor(y_test_seq, dtype=torch.float32)
)

train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True)
val_loader   = DataLoader(val_ds,   batch_size=batch_size, shuffle=False)
test_loader  = DataLoader(test_ds,  batch_size=batch_size, shuffle=False)

# ====================================================
# 6. LSTM Model
# ====================================================
class LSTMMulti(nn.Module):
    def __init__(self, num_features, hidden_size=64):
        super().__init__()
        self.lstm = nn.LSTM(
            input_size=num_features,
            hidden_size=hidden_size,
            batch_first=True
        )
        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        out, _ = self.lstm(x)     # out: (batch, seq_len, hidden)
        out = out[:, -1, :]       # last time step
        out = self.fc(out)        # (batch, 1)
        return out

num_features = X_train_seq.shape[2]
model = LSTMMulti(num_features).to(device)
print(model)

# ====================================================
# 7. Training
# ====================================================
loss_fn = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

EPOCHS = 10

for epoch in range(1, EPOCHS + 1):
    model.train()
    train_losses = []
    for xb, yb in train_loader:
        xb, yb = xb.to(device), yb.to(device).unsqueeze(1)

        optimizer.zero_grad()
        pred = model(xb)
        loss = loss_fn(pred, yb)
        loss.backward()
        optimizer.step()
        train_losses.append(loss.item())

    # Validation
    model.eval()
    val_losses = []
    with torch.no_grad():
        for xb, yb in val_loader:
            xb, yb = xb.to(device), yb.to(device).unsqueeze(1)
            pred = model(xb)
            val_losses.append(loss_fn(pred, yb).item())

    print(f"Epoch {epoch:02d} | Train Loss={np.mean(train_losses):.6f} | Val Loss={np.mean(val_losses):.6f}")

# ====================================================
# 8. Evaluation on test set
# ====================================================
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

model.eval()
y_pred_scaled = []
y_true_scaled = []

with torch.no_grad():
    for xb, yb in test_loader:
        xb = xb.to(device)
        pred = model(xb).cpu().numpy()
        y_pred_scaled.append(pred.reshape(-1))
        y_true_scaled.append(yb.numpy())

y_pred_scaled = np.concatenate(y_pred_scaled)
y_true_scaled = np.concatenate(y_true_scaled)

# inverse scaling
y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1)).flatten()
y_true = scaler_y.inverse_transform(y_true_scaled.reshape(-1, 1)).flatten()

print("\n[Test] Multi-feature LSTM performance:")
print("RMSE =", mean_squared_error(y_true, y_pred, squared=False))
print("MAE  =", mean_absolute_error(y_true, y_pred))
print("R^2  =", r2_score(y_true, y_pred))
```

```python
from captum.attr import IntegratedGradients
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
model.eval()  # 确保在 eval 模式

def forward_func(x):
    """
    x: tensor of shape (batch, seq_len, num_features)
    returns: tensor of shape (batch,) or (batch, 1)
    """
    out = model(x)          # (batch, 1)
    return out.squeeze(-1)  # (batch,)
```

```python
# 选择一个样本
sample_id = 100
x_sample = X_test_seq[sample_id:sample_id+1]  # shape: (1, seq_len, D)

# 转成 tensor，放到 device 上
x_input = torch.tensor(x_sample, dtype=torch.float32, device=device)

# 定义 baseline（参照输入）
# 这里用全 0 baseline，你也可以改成用训练集特征均值填充
baseline = torch.zeros_like(x_input)

# 初始化 Integrated Gradients
ig = IntegratedGradients(forward_func)

# 计算 IG 归因
# n_steps 越大越精细，但也越慢，可按需要调整
attributions, delta = ig.attribute(
    x_input,
    baselines=baseline,
    n_steps=50,
    return_convergence_delta=True
)

print("Attributions shape:", attributions.shape)  # (1, seq_len, D)
print("Convergence delta:", delta)

# 转成 numpy，取绝对值看重要性
attr_np = attributions[0].detach().cpu().numpy()   # (seq_len, D)
attr_abs = np.abs(attr_np)
```

```python
# 取测试集前 N 个样本做全局分析
N_SAMPLES = 500
X_batch = X_test_seq[:N_SAMPLES]  # (N_SAMPLES, seq_len, D)

x_input_batch = torch.tensor(X_batch, dtype=torch.float32, device=device)
baseline_batch = torch.zeros_like(x_input_batch)

ig = IntegratedGradients(forward_func)

# 对一整个 batch 做 IG
attributions_batch, delta_batch = ig.attribute(
    x_input_batch,
    baselines=baseline_batch,
    n_steps=50,
    return_convergence_delta=True
)

attr_batch_np = attributions_batch.detach().cpu().numpy()  # (N_SAMPLES, seq_len, D)
attr_batch_abs = np.abs(attr_batch_np)

# 在样本维度 + 时间维度求平均 -> 每个特征一个重要性
feature_importance = attr_batch_abs.mean(axis=(0, 1))      # shape: (D,)

# 排序并打印
idx_sorted = np.argsort(-feature_importance)

print("\nGlobal feature importance (Integrated Gradients):")
for i in idx_sorted:
    print(f"{FEATURES[i]:<20s}: {feature_importance[i]:.6e}")

# 画条形图
plt.figure(figsize=(8, 5))
plt.bar(range(len(FEATURES)), feature_importance[idx_sorted])
plt.xticks(range(len(FEATURES)), [FEATURES[i] for i in idx_sorted], rotation=90)
plt.ylabel("Average |IG|")
plt.title("Global feature importance (Integrated Gradients)")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches

# ---- 1. 定义特征类别，用来上色 ----
def get_feature_category(name):
    if name.startswith("CO2_rollmean"):
        return "rollmean"
    elif name.startswith("CO2_rollvar"):
        return "rollvar"
    elif name.startswith("CO2_lag"):
        return "lag"
    elif name in ["hour", "dayofweek", "month"]:
        return "time"
    else:
        return "other"

color_map = {
    "rollmean": plt.cm.Blues(0.7),
    "rollvar":  plt.cm.Greens(0.6),
    "lag":      plt.cm.Oranges(0.6),
    "time":     plt.cm.Purples(0.6),
    "other":    "grey",
}

# ---- 2. 按重要性从大到小排序 ----
sorted_features = [FEATURES[i] for i in idx_sorted]          # 大->小
sorted_values   = feature_importance[idx_sorted]             # 大->小

# y 轴位置：为了让“最大的在最上面”，用反向序列
# 例如有 5 个特征，则 y_pos = [4,3,2,1,0]
y_pos = np.arange(len(sorted_features))[::-1]

# 颜色按类别映射（顺序与 sorted_features 一致）
bar_colors = [color_map[get_feature_category(f)] for f in sorted_features]

# ---- 3. 画横向条形图 ----
plt.figure(figsize=(8, 8))

plt.barh(
    y_pos,              # 控制谁在上谁在下
    sorted_values,      # 从大到小
    color=bar_colors
)

# y 轴标签放在对应的 y_pos 上
plt.yticks(y_pos, sorted_features, fontsize=9)

plt.xlabel("Average |IG|", fontsize=12)
plt.title("Global Feature Importance (Integrated Gradients)", fontsize=14)

# 去掉上、右边框
ax = plt.gca()
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

# 图例
legend_elements = [
    mpatches.Patch(color=color_map["rollmean"], label="Rolling Mean"),
    mpatches.Patch(color=color_map["rollvar"],  label="Rolling Variance"),
    mpatches.Patch(color=color_map["lag"],      label="Lag Features"),
    mpatches.Patch(color=color_map["time"],     label="Time Features"),
]
plt.legend(handles=legend_elements, title="Feature Groups", loc="lower right")

plt.tight_layout()
plt.show()
```
